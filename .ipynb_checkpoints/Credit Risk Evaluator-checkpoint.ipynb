# Import Dependencies

```python
import numpy as np
import pandas as pd
from pathlib import Path
```

# Use 2019 data to test against 2020 Data

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))

train_df
```

# Assign X and y train data

```python
X_train = pd.get_dummies(train_df.drop(columns=['target']))
y_train = train_df['target']
```

```python
X_train.head()
```

# Assign X and y test data

```python
X_test = pd.get_dummies(test_df.drop(columns=['target']))
y_test = test_df['target']
```

# Add missing dummy variables to testing set

```python
for col in X_train.columns:
    if col not in X_test.columns:
        X_test[col] = 0
```

# Prediction before moving forward with Logistic Regression and Random Forest models

I think the Random Forest model will perform better than Logistic Regression because there are so many features to consider. The Random Forest model will use many decision trees, so I think that will result in a better outcome for the data supplied. 

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score

clf = LogisticRegression(solver='lbfgs', random_state=1)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Train a Random Forest Classifier model and print the model score

clf = RandomForestClassifier()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# Scale the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

As expected, the Random Forest model performed better than the logistic Regression model. Random Forest was 64.65%.
Much better than the 50.83% of the Logistic Regression model. However, both scores are pretty low. 

I predect that scaling will have a positive impact on the logistic regression score but not really change the Random Forest accuracy as it's a tree-based model.

# Logistic Regression with Scaled Data

```python
# Train the Logistic Regression model on the scaled data and print the model score

from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver='lbfgs')
clf.fit(X_train_scaled, y_train)
clf.score(X_test_scaled, y_test)
```

# Random Forest with Scaled Data

```python
# Train a Random Forest Classifier model on the scaled data and print the model score

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(X_train_scaled, y_train)
clf.score(X_test_scaled, y_test)
```

# Final Comments

I predicted that the Logistic Regression accuracy would improve with scaling and that the Random Forest would not change much with scaling. I was correct. The Random Forest models (scaled and unscaled) stayed about the same. The Logistic Regression model with scaled data was the best performing overall.
